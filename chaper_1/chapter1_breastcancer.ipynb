```python
import pandas as pd
import numpy as np
file_path='data/BreastCancer/breast-cancer-wisconsin.data.txt'
data=pd.read_csv(file_path, delimiter=',',header =None,skipinitialspace=True)
#print(data)
```

```python
data.head(20)
#data.describe()
#data.info()
```

```python
data=data.replace('?', np.NaN)
data.info()
```

```python
data=data.dropna()
#data=data.replace(np.NaN, 1)
#data = data.apply(lambda x:x.fillna(x.value_counts().index[0]))
data.info()
#data.iloc[:,6]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test,y_train,y_test = train_test_split(data.iloc[:,1:10], data.iloc[:,10],test_size=0.3, random_state=41)

print(x_train.shape)
#y_train
```

```python
from sklearn.neighbors import KNeighborsClassifier

neigh = KNeighborsClassifier(n_neighbors=9)
neigh.fit(x_train, y_train)
```

```python
y_pred=neigh.predict(x_test)

print(np.asarray(y_test[:30]))

print(y_pred[:30])
```

```python
from sklearn import metrics
score_test = metrics.accuracy_score(y_test, y_pred)
print(score_test)
```
